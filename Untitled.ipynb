## 1. Data Quality Checks
**Null or missing values check**

As part of our quality checks, let us quick look at whether there are any null values in the train dataset as follows:

```python
# any() applied twice to check run the isnull check across all columns
train.isnull().any().any()
```

- Our null values check returns False but however, this does not really mean that this case has been closed as the data is also described as "Values of -1 indicate that the feature was missing from the observation". Therefore I take it that Porto Seguro has simply conducted a blanket replacement of all null values in the data with the value of -1. Let us now inspect if there are any missing values in the data

Here we can see that which columns contained -1 in their values so we could easily for example make a blanket replacement of all -1 with nulls first as follows:

```python
train_copy = train
train_copy = train_copy.replace(-1, np.NaN)
```

```python
import missingno as msno

# Nullity or mising values by columns
msno.matrix(df=train_copy.iloc[:,2:39], figsize=(20,14), color=(0.42, 0.1, 0.05))
```

Another standard check normally conducted on the data is with regards to our target variable, where in this case, the column is conveniently titled "target". The target value also comes by the moniker

```python
data = [go.Bar(
    x = train['target'].value_counts().index.values,
    y = train['target'].value_counts().values,
    text = 'Distribution of target variable'
)]

layout = go.Layout(
    title = 'Target variable distribution'
)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig, filename='basic-bar')
```
